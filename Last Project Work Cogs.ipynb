# Step 1: Importing Libraries and Reading Data
* - Import necessary libraries: pandas, numpy, xlrd, and xlsxwriter for data manipulation, numerical operations, and working with Excel files.
* - Read data from three Excel files into DataFrames df5, df6, and df7.



```python
# Step 1: Importing Libraries and Reading Data
import pandas as pd
import numpy as np
import xlrd
import xlsxwriter
```

```python


# Explanation: Importing necessary libraries and reading data from three Excel files into DataFrames
df5 = pd.read_excel('Gesamt-Kauf.xlsx')
df6 = pd.read_excel('Inventur.xlsx')
df7 = pd.read_excel('Weste.xlsx')

```

# Step 2: Merging DataFrames
* - First merge: Merge df5 and df6 on the column 'ArtikelNr' using an inner join to keep only matching rows.
* - Second merge: Merge the resulting DataFrame (cogs) with df7 on the column 'ArtikelNr', adding suffixes to differentiate similar column names from df7.



```python

# First merge: df5 with df6 on 'ArtikelNr'
cogs = pd.merge(df5, df6, on='ArtikelNr', how='inner')

# Second merge: cogs with df7 on 'ArtikelNr', suffixes added to avoid column name conflicts
cogs = pd.merge(cogs, df7, on='ArtikelNr', how='inner', suffixes=('', '_df7'))

# Print or further process the merged DataFrame 'cogs'
cogs


```


# Step 7: Visualizing Data
* Top 5 and Bottom 5 Articles by Quantity:
* - Sort and select the top 5 articles by 'Anzahl'.
* - Sort and select the bottom 5 articles by 'Anzahl', excluding entries with 0.
* - Set color palettes for the pie charts.
* - Create pie charts to visualize the top and bottom 5 articles by quantity.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure ArtikelName is treated as a string (categorical data)
cogs['ArtikelName'] = cogs['ArtikelName'].astype(str)

# Filter out entries where ArtikelName is '0'
cogs = cogs[cogs['ArtikelName'] != '0']

# Sort the dataframe by Anzahl in descending order to get the top 5 articles
top_5_articles = cogs.sort_values(by='Anzahl', ascending=False).head(5)

# Sort the dataframe by Anzahl in ascending order to get the bottom 5 articles
# Ensure articles with 0 Anzahl are not included
bottom_5_articles = cogs[cogs['Anzahl'] > 0].sort_values(by='Anzahl', ascending=True).head(5)

# Set a color palette
top_palette = sns.color_palette("Greens", len(top_5_articles))
bottom_palette = sns.color_palette("Reds", len(bottom_5_articles))

# Plotting the top 5 articles by Anzahl
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))

# Top 5 articles pie chart
axes[0].pie(top_5_articles['Anzahl'], labels=top_5_articles['ArtikelName'], autopct='%1.1f%%', colors=top_palette, startangle=140)
axes[0].set_title('Top 5 Group of Product that are boughtthe most')

# Bottom 5 articles pie chart
axes[1].pie(bottom_5_articles['Anzahl'], labels=bottom_5_articles['ArtikelName'], autopct='%1.1f%%', colors=bottom_palette, startangle=140)
axes[1].set_title('Bottom 5 Group of product that are bought the least')

plt.tight_layout()
plt.show()
```


# Top 5 and Bottom 5 Articles by Cost:
* - Sort and select the top 5 articles by 'EUR'.
* - Sort and select the bottom 5 articles by 'EUR', excluding entries with 0.
* - Set color palettes for the pie charts.
* - Create pie charts to visualize the top and bottom 5 articles by cost.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure VollName_x is treated as a string (categorical data)
cogs['VollName_x'] = cogs['VollName_x'].astype(str)

# Filter out entries where VollName_x is '0'
cogs = cogs[cogs['VollName_x'] != '0']

# Sort the dataframe by EUR in descending order to get the top 5 articles by cost
top_5_cost_articles = cogs.sort_values(by='EUR', ascending=False).head(5)

# Sort the dataframe by EUR in ascending order to get the bottom 5 articles by cost
# Ensure articles with 0 EUR are not included
bottom_5_cost_articles = cogs[cogs['EUR'] > 0].sort_values(by='EUR', ascending=True).head(5)

# Set a color palette
top_palette = sns.color_palette("Blues", len(top_5_cost_articles))
bottom_palette = sns.color_palette("Reds", len(bottom_5_cost_articles))

# Plotting the top 5 articles by cost
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))

# Top 5 articles pie chart
axes[0].pie(top_5_cost_articles['EUR'], labels=top_5_cost_articles['VollName_x'], autopct='%1.1f%%', colors=top_palette, startangle=140)
axes[0].set_title('Top 5 Costly Product that are bought (EUR)')

# Bottom 5 articles pie chart
axes[1].pie(bottom_5_cost_articles['EUR'], labels=bottom_5_cost_articles['VollName_x'], autopct='%1.1f%%', colors=bottom_palette, startangle=140)
axes[1].set_title('Bottom 5 Least costly that are bought (EUR)')

plt.tight_layout()
plt.show()
```
