````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load the data from Excel files
df1 = pd.read_excel("Jan-Feb.xlsx")
df2 = pd.read_excel("Marz-April.xlsx")
df3 = pd.read_excel("Mai-Jun.xlsx")
df4 = pd.read_excel("Jul-August.xlsx")
df5 = pd.read_excel("Sep-Okt.xlsx")
df6 = pd.read_excel("Nov-Dez.xlsx")
````

````python

# Merge all the DataFrames on 'Artikelname'
Daily_sales = pd.merge(df1, df2, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df3, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df4, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df5, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df6, on='Artikelname', how='inner')
````

````python
# Fill NaN values with 0
Daily_sales = Daily_sales.fillna(value=0)
````

````python
# Print data types and a summary of the data
Daily_sales.dtypes
````

````python
Daily_sales.describe()
````

````python
# Calculate the total sales for each product
Daily_sales['Total_sales'] = Daily_sales.iloc[:, 1:].sum(axis=1)
````

````python
# Calculate the total number of products sold
total_products_sold = Daily_sales.iloc[:, 1:].sum().sum()
print("Total products sold:", total_products_sold)
````


# EDA Explanation:
# 1. **Loading and Merging Data**: Load data from multiple Excel files and merge them into a single DataFrame.
# 2. **Handling Missing Values**: Fill missing values with 0.
# 3. **Summary Statistics**: Print data types and descriptive statistics for the DataFrame.
# 4. **Calculating Sales**: Calculate total sales for each product and the overall total sales.

````python
# Setting a seaborn style
sns.set(style="whitegrid")

# Top 10 Selling Products
top_10_products = Daily_sales.sort_values(by='Total_sales', ascending=False).head(10)
print("Total sales for top 10 selling products:")
print(top_10_products[['Artikelname', 'Total_sales']])

# Plotting the sales of the top 10 selling products
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=10)
barplot = plt.bar(top_10_products['Artikelname'], top_10_products['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Top 10 Selling Products of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python


# Bottom 10 Selling Products
bottom_10_products = Daily_sales.sort_values(by='Total_sales', ascending=True).head(10)
print("Total sales for bottom 10 selling products:")
print(bottom_10_products[['Artikelname', 'Total_sales']])

# Plotting the sales of the bottom 10 selling products
plt.figure(figsize=(12, 8))
colors = sns.color_palette('coolwarm', n_colors=10)
barplot = plt.bar(bottom_10_products['Artikelname'], bottom_10_products['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Bottom 10 Selling Products of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python

# Analysis for Core Artikel
Core_Artikel = Daily_sales[Daily_sales['Artikelname'].str.contains('Kater', case=False, regex=True)]

# Drop specific articles from Core Artikel
Core_Artikel = Core_Artikel[~Core_Artikel['Artikelname'].isin(['Katerkugel Apfel-Zimt', 'Katerkugel Schoko-Kirsch'])]

# Calculate total core products sold
total_core_products_sold = Core_Artikel.iloc[:, 1:].sum().sum()
print("Total core products sold:", total_core_products_sold)

# Top 5 Core Artikel
Core_Artikel['Total_sales'] = Core_Artikel.iloc[:, 1:].sum(axis=1)
top_5_core_artikel = Core_Artikel.sort_values(by='Total_sales', ascending=False).head(5)
print("Total sales for top 5 core articles:")
print(top_5_core_artikel[['Artikelname', 'Total_sales']])

# Plotting the sales of the top 5 Core Artikel
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=5)
barplot = plt.bar(top_5_core_artikel['Artikelname'], top_5_core_artikel['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Artikelname')
plt.ylabel('Total Sales')
plt.title('Top 5 Core Artikel of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Bottom 5 Core Artikel
min_5_core_artikel = Core_Artikel.sort_values(by='Total_sales', ascending=True).head(5)
print("Total sales for bottom 5 core articles:")
print(min_5_core_artikel[['Artikelname', 'Total_sales']])

# Plotting the sales of the bottom 5 Core Artikel
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=5)
barplot = plt.bar(min_5_core_artikel['Artikelname'], min_5_core_artikel['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Artikelname')
plt.ylabel('Total Sales')
plt.title('Bottom 5 Core Artikel of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python

# Analysis for Bowls
Bowl = Daily_sales[Daily_sales['Artikelname'].str.contains('Bowl', case=False, regex=True)]

# Calculate total Bowl products sold
total_Bowl_sold = Bowl.iloc[:, 1:].sum().sum()
print("Total Bowl sold:", total_Bowl_sold)

# Top 5 Bowl Artikel
Bowl['Total_sales'] = Bowl.iloc[:, 1:].sum(axis=1)
top_5_bowl_artikel = Bowl.sort_values(by='Total_sales', ascending=False).head(5)
print("Total sales for top 5 Bowl articles:")
print(top_5_bowl_artikel[['Artikelname', 'Total_sales']])

# Plotting the sales of the top 5 Bowl Artikel
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=5)
barplot = plt.bar(top_5_bowl_artikel['Artikelname'], top_5_bowl_artikel['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Artikelname')
plt.ylabel('Total Sales')
plt.title('Top 5 Bowl Artikel of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Bottom 5 Bowl Artikel
min_5_bowl_artikel = Bowl.sort_values(by='Total_sales', ascending=True).head(5)
print("Total sales for bottom 5 Bowl articles:")
print(min_5_bowl_artikel[['Artikelname', 'Total_sales']])

# Plotting the sales of the bottom 5 Bowl Artikel
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=5)
barplot = plt.bar(min_5_bowl_artikel['Artikelname'], min_5_bowl_artikel['Total_sales'], color=colors)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.xlabel('Artikelname')
plt.ylabel('Total Sales')
plt.title('Bottom 5 Bowl Artikel of 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Analyzing Daily Totals for Bowl
daily_totals = Bowl.iloc[:, 1:].sum(axis=0)

# Find the maximum and minimum daily totals along with their dates
max_daily_total = daily_totals.max()
min_daily_total = daily_totals.min()
max_daily_total_date = daily_totals.idxmax()
min_daily_total_date = daily_totals.idxmin()

# Drop rows where values are 0 on the max and min total dates
filtered_max_date = Bowl[Bowl[max_daily_total_date] != 0][['Artikelname', max_daily_total_date]]
filtered_min_date = Bowl[Bowl[min_daily_total_date] != 0][['Artikelname', min_daily_total_date]]

# Plotting the sales for the date with the maximum daily total using a horizontal bar plot
plt.figure(figsize=(18, 10))
colors = plt.cm.viridis(np.linspace(0, 1, len(filtered_max_date)))
bars = plt.barh(filtered_max_date['Artikelname'], filtered_max_date[max_daily_total_date], color=colors)
plt.title(f'Bowl Sales Distribution on {max_daily_total_date} (Maximum Total)', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Artikelname')
total_sales_max = filtered_max_date[max_daily_total_date].sum()
for bar in bars:
    width = bar.get_width()
    percentage = (width / total_sales_max) * 100
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', va='center', ha='left', fontsize=14)
plt.show()
````

````python

# Plotting the sales for the date with the minimum daily total using a horizontal bar plot
plt.figure(figsize=(18, 5))
colors = plt.cm.viridis(np.linspace(0, 1, len(filtered_min_date)))
bars = plt.barh(filtered_min_date['Artikelname'], filtered_min_date[min_daily_total_date], color=colors)
plt.title(f'Bowl Sales Distribution on {min_daily_total_date} (Minimum Total)', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Artikelname')
total_sales_min = filtered_min_date[min_daily_total_date].sum()
for bar in bars:
    width = bar.get_width()
    percentage = (width / total_sales_min) * 100
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', va='center', ha='left', fontsize=14)
plt.show()
````

````python

# Minimum Daily Total for Bowls
plt.subplot(3, 2, 6)
bars = plt.barh(filtered_min_date['Artikelname'], filtered_min_date[min_daily_total_date], color=colors)
plt.title(f'Sales Distribution on {min_daily_total_date} (Minimum Total)', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Artikelname')
total_sales_min = filtered_min_date[min_daily_total_date].sum()
for bar in bars:
    width = bar.get_width()
    percentage = (width / total_sales_min) * 100
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', va='center', ha='left', fontsize=14)

plt.tight_layout()
plt.show()
````
