# Daily Product that are sold in year 2023 






## Step 1: Importing Libraries and Reading Data
* - Import necessary libraries: pandas, numpy, os, xlrd, xlsxwriter, matplotlib, seaborn.
* - Read data from six Excel files into DataFrames (df1 to df6).


```python
import pandas as pd
import numpy as np
import os
import os.path
import xlrd
import xlsxwriter
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df1 = pd.read_excel("Jan-Feb.xlsx")
df2 = pd.read_excel("Marz-April.xlsx")
df3 = pd.read_excel("Mai-Jun.xlsx")
df4 = pd.read_excel("Jul-August.xlsx")
df5 = pd.read_excel("Sep-Okt.xlsx")
df6 = pd.read_excel("Nov-Dez.xlsx")
```


* Result: Six DataFrames containing sales data from different months.
* Step 2: Merging DataFrames


* - Merge all six DataFrames (df1 to df6) on the column 'Artikelname' using inner joins to combine daily sales data into a single DataFrame 'Daily_sales'.


```python
Daily_sales = pd.merge(df1, df2, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df3, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df4, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df5, on='Artikelname', how='inner')
Daily_sales = pd.merge(Daily_sales, df6, on='Artikelname', how='inner')

print(Daily_sales)
```


* Result: A combined DataFrame 'Daily_sales' with 110 rows and 366 columns, each representing daily sales for different articles.

### Step 8: Identify Top 10 Selling Products
##### The dataframe is sorted by 'Total_sales' in descending order, and the top 10 products are selected. 
##### The total sales of these top 10 products are printed, highlighting the best-selling items.

```python
import seaborn as sns

# Setting a seaborn style
sns.set(style="whitegrid")

# Assuming 'Daily_sales' is your DataFrame
# Summing across all date columns for each product
Daily_sales['Total_sales'] = Daily_sales.iloc[:, 1:].sum(axis=1)  # Sum from the second column onwards

# Sorting the DataFrame by 'Total_sales' in descending order and selecting the top 10
top_10_products = Daily_sales.sort_values(by='Total_sales', ascending=False).head(10)

# Printing the total sales of these top 10 products
print("Total sales for top 10 selling products:", top_10_products[['Artikelname', 'Total_sales']])

# Plotting the sales of the top 10 selling products
plt.figure(figsize=(12, 8))
colors = sns.color_palette('husl', n_colors=10)  # A more colorful, distinct palette
barplot = plt.bar(top_10_products['Artikelname'], top_10_products['Total_sales'], color=colors)

# Adding data labels on each bar
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')

plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Top 10 Selling Products of 2023')
plt.xticks(rotation=45)  # Rotates the product names for better visibility
plt.tight_layout()  # Adjusts plot parameters to give some padding and prevent overlap
plt.show()

```

#### Doing this in new Plot, Sunburst

```python
import plotly.graph_objects as go
import pandas as pd

# Assuming Daily_sales is a predefined DataFrame
daily_totals = Daily_sales.iloc[:, 1:].sum(axis=0)

# Find the date with the maximum daily total
max_daily_total_date = daily_totals.idxmax()

# Drop rows where values are 0 on the max total date
filtered_max_date = Daily_sales[Daily_sales[max_daily_total_date] != 0][['Artikelname', max_daily_total_date]]

# Sort the data by sales values in descending order
filtered_max_date = filtered_max_date.sort_values(by=max_daily_total_date, ascending=False)

# Creating the hierarchical structure for the Sunburst chart
labels = filtered_max_date['Artikelname'].tolist()
values = filtered_max_date[max_daily_total_date].tolist()

# Create ids and parents lists for the Sunburst chart
ids = ["Total Sales"] + labels
parents = [""] + ["Total Sales"] * len(labels)

# Add the total sales value
values_total = [sum(values)] + values

# Create the Sunburst chart object
fig = go.Figure()

fig.add_trace(go.Sunburst(
    ids=ids,
    labels=ids,
    parents=parents,
    values=values_total,
    domain=dict(column=1),
    maxdepth=2,
    insidetextorientation='radial'
))

fig.update_layout(
    title_text=f"Sales Distribution on {max_daily_total_date} (Max Total)",
    margin=dict(t=40, l=10, r=10, b=10),
    width=800,
    height=800
)

fig.show()


```

```python
import holoviews as hv
from holoviews import opts
import pandas as pd
hv.extension('bokeh')

# Assuming daily_totals and Daily_sales are predefined DataFrames

# Find the date with the minimum daily total
min_daily_total_date = daily_totals.idxmin()

# Drop rows where values are 0 on the min total date
filtered_min_date = Daily_sales[Daily_sales[min_daily_total_date] != 0][['Artikelname', min_daily_total_date]]

# Creating a DataFrame suitable for a Chord diagram
chord_data = filtered_min_date.rename(columns={min_daily_total_date: 'value'})
chord_data['source'] = chord_data['Artikelname']
chord_data['target'] = chord_data['Artikelname']
chord_data = chord_data[['source', 'target', 'value']]

# Create nodes dataset
nodes = pd.DataFrame({'index': range(len(chord_data)), 'name': chord_data['source']})
nodes_dataset = hv.Dataset(nodes, 'index')

# Create links dataset
links = chord_data[['source', 'target', 'value']]
links = links.merge(nodes[['index', 'name']], left_on='source', right_on='name')
links = links.merge(nodes[['index', 'name']], left_on='target', right_on='name', suffixes=('_source', '_target'))
links = links[['index_source', 'index_target', 'value']].rename(columns={'index_source': 'source', 'index_target': 'target'})

# Creating a Chord object
chord = hv.Chord((links, nodes_dataset)).select(value=(0, None))

# Styling the Chord diagram with a title
chord.opts(
    opts.Chord(
        cmap='Category20', edge_color=hv.dim('source').str(), labels='name', node_color=hv.dim('index').str(),
        label_text_font_size='10pt', width=1000, height=800, 
        title=f'Sales Distribution on {min_daily_total_date} (Min Total)'
    )
)

# Display the plot
hv.output(chord)
```

```python

# })

# Select only the numeric columns for summing (all columns except 'Artikelname')
daily_totals = Bowl.iloc[:, 1:].sum(axis=0)

# Find the maximum and minimum daily totals along with their dates
max_daily_total = daily_totals.max()
min_daily_total = daily_totals.min()
max_daily_total_date = daily_totals.idxmax()
min_daily_total_date = daily_totals.idxmin()

# Drop rows where values are 0 on the max and min total dates
filtered_max_date = Bowl[Bowl[max_daily_total_date] != 0][['Artikelname', max_daily_total_date]]
filtered_min_date = Bowl[Bowl[min_daily_total_date] != 0][['Artikelname', min_daily_total_date]]

# Plotting the sales for the date with the maximum daily total using a horizontal bar plot
plt.figure(figsize=(18, 10))
colors = plt.cm.viridis(np.linspace(0, 1, len(filtered_max_date)))
bars = plt.barh(filtered_max_date['Artikelname'], filtered_max_date[max_daily_total_date], color=colors)
plt.title(f'Bowl Sales Distribution on {max_daily_total_date} (Maximum Total)', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Artikelname')

# Adding percentage values at the end of each bar
total_sales_max = filtered_max_date[max_daily_total_date].sum()
for bar in bars:
    width = bar.get_width()
    percentage = (width / total_sales_max) * 100
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', va='center', ha='left', fontsize=14)

plt.show()

# Plotting the sales for the date with the minimum daily total using a horizontal bar plot
plt.figure(figsize=(18, 10))
colors = plt.cm.viridis(np.linspace(0, 1, len(filtered_min_date)))
bars = plt.barh(filtered_min_date['Artikelname'], filtered_min_date[min_daily_total_date], color=colors)
plt.title(f'Bowl Sales Distribution on {min_daily_total_date} (Minimum Total)', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Artikelname')

# Adding percentage values at the end of each bar
total_sales_min = filtered_min_date[min_daily_total_date].sum()
for bar in bars:
    width = bar.get_width()
    percentage = (width / total_sales_min) * 100
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{percentage:.1f}%', va='center', ha='left', fontsize=14)

plt.show()

```
